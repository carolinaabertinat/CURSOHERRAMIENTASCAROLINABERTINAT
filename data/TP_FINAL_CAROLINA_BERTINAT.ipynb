#EXPLORACION Y VISUALIZACIÓN

- GRAFICO DE SALARIO POR ROL:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the DataFrame first, as it was not defined
df = pd.read_csv('/content/drive/MyDrive/CURSOHERRAMIENTAS/AI Job Market Dataset.csv')

plt.figure(figsize=(10,5))
sns.barplot(data=df, x="job_title", y="salary") # Changed 'salary_in_usd' to 'salary'
plt.xticks(rotation=45)
plt.title("Salario promedio por rol")
plt.show()
```

- SALARIO VS. EXXPERIENCIA:

```python
sns.boxplot(data=df, x="experience_level", y="salary")
plt.title("Salario según experiencia")
plt.show()
```

- DISTRIBUCIÓN DE SALARIOS:

```python
sns.histplot(df['salary'], bins=30, kde=True)
plt.title("Distribución de salarios")
plt.show()
```

- CANTIDAD DE EMPLEOS SEGUN EL PAIS:

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top países con más empleos")
plt.show()
```
